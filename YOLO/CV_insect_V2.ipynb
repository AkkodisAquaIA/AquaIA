```python
import cv2
```

```python
#Load the trained model (best weight)
model = YOLO("C:/Users/agniraj.senathipathi/Documents/Project/runs/detect/train7/weights/best.pt")
```

```python
# Load the image
image_path = "C:/Users/agniraj.senathipathi/Documents/Project/Test_insect.jpg"
#image_path = "C:/Users/agniraj.senathipathi/Documents/Aqua_IA/Practice/Animal_Data/valid/images/4fadd04098.jpg"
image = cv2.imread(image_path)
```

```python
# Run prediction on an image
results = model.predict(image_path, save=True, conf=0.25)
```

```python
# Display the image with predictions (optional)
for r in results:
    # Plot results using OpenCV
    annotated_frame = r.plot()
    cv2.imshow("Predictions", annotated_frame)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# Create output directory for cropped objects
import os
output_dir = "cropped_objects"
os.makedirs(output_dir, exist_ok=True)
```

```python
# Process results
for i, result in enumerate(results):
    boxes = result.boxes
    for j, box in enumerate(boxes):
        # Get bounding box coordinates
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # top-left and bottom-right
        cropped_img = image[y1:y2, x1:x2]

        # Optional: Get class name
        cls_id = int(box.cls[0])
        cls_name = model.names[cls_id]

        # Save the cropped image
        filename = f"{output_dir}/object_{i}_{j}_{cls_name}.jpg"
        cv2.imwrite(filename, cropped_img)

        print(f"Saved: {filename}")
```
